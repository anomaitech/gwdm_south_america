```python
wells=pd.read_excel('/content/Well Name Latitude Longitude.xlsx')
```

```python
wells['Well ID'] = range(1, len(wells) + 1)
wells['Aquifer ID']=1
display(wells)
```

```python
# Create a mapping from well names in merged_erin to Well IDs in wells
well_name_to_id = wells[['Well Name', 'Well ID','Aquifer ID']].set_index('Well Name').to_dict()['Well ID']

# Melt the merged_erin DataFrame to long format
merged_erin_long = merged_erin.melt(id_vars=['Date'], var_name='Well Name', value_name='Value')

# Map the Well Name in the long format DataFrame to Well ID
merged_erin_long['Well ID'] = merged_erin_long['Well Name'].map(well_name_to_id)

# Display the head of the resulting DataFrame
display(merged_erin_long.head())
```

```python
merged_erin_long['Date']=pd.to_datetime(merged_erin_long['Date'])
```
